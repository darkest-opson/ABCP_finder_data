```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io

data = """Model ACC SP PRE SE F1 MCC AUC
LR 0.9086 0.9228 0.8733 0.8851 0.8792 0.8058 0.9559
RF 0.9010 0.9472 0.9037 0.8243 0.8622 0.7872 0.9551
SVC 0.9112 0.9228 0.8742 0.8919 0.8829 0.8115 0.9485
KNN 0.9036 0.9309 0.8819 0.8581 0.8699 0.7935 0.9493
GB 0.8985 0.9350 0.8857 0.8378 0.8611 0.7820 0.9532
AB 0.8706 0.9146 0.8489 0.7973 0.8223 0.7215 0.9467
XGB 0.8985 0.9228 0.8699 0.8581 0.8639 0.7830 0.9560
MLP 0.8909 0.9350 0.8832 0.8176 0.8491 0.7652 0.9447
Ensemble 0.9061 0.9309 0.8828 0.8649 0.8737 0.7991 0.9612"""


data_lines = data.strip().split('\n')
headers = data_lines[0].split()
rows = []

for line in data_lines[1:]:
    parts = line.split()
    model_name_parts = []
    for part in parts:
        try:
            float(part)
            break
        except ValueError:
            model_name_parts.append(part)
    model_name = ' '.join(model_name_parts)
    metrics = [float(p) for p in parts[len(model_name_parts):]]
    rows.append([model_name] + metrics)

df = pd.DataFrame(rows, columns=['Model', 'ACC', 'SP', 'PRE', 'SE', 'F1', 'MCC', 'AUC'])

df_plot = df.set_index('Model')
metrics = df_plot.columns
models = df_plot.index

fig, ax = plt.subplots(figsize=(15, 8))

n_metrics = len(metrics)
n_models = len(models)
bar_width = 0.8 / n_metrics
index = np.arange(n_models)

# Define a more distinct colormap
colors = plt.cm.get_cmap('tab10', n_metrics) # 'tab10' provides 10 distinct colors

for i, metric in enumerate(metrics):
    bars = ax.bar(index + i * bar_width, df_plot[metric], bar_width, label=metric, color=colors(i))
    # Add values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom', fontsize=8, rotation=90)


ax.set_xlabel('Machine Learning Models', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Performance Metrics of Different Machine Learning Models Trained on ProtBERT embeddings', fontsize=10)
ax.set_xticks(index + 0.4)
ax.set_xticklabels(models, rotation=0, ha='right')
ax.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')

ax.set_ylim(0, 1.1)
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.savefig('ProtBERT_ML_img.png',dpi=300, bbox_inches="tight")
```
