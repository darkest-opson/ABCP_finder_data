```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_esm = pd.read_csv("model_performance_ESM.csv")
df_esm
```

```python
df_plot = df_esm.set_index('Model')
metrics = df_plot.columns
models = df_plot.index

fig, ax = plt.subplots(figsize=(15, 8))

n_metrics = len(metrics)
n_models = len(models)
bar_width = 0.8 / n_metrics
index = np.arange(n_models)

# Define a more distinct colormap
colors = plt.cm.get_cmap('tab10', n_metrics) # 'tab10' provides 10 distinct colors

for i, metric in enumerate(metrics):
    bars = ax.bar(index + i * bar_width, df_plot[metric], bar_width, label=metric, color=colors(i))
    # Add values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom', fontsize=8, rotation=90)


ax.set_xlabel('', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('', fontsize=10)
ax.set_xticks(index + 0.4)
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')

ax.set_ylim(0, 1.1)
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.savefig('ESM2_ML_img.png',dpi=600, bbox_inches="tight")
```
